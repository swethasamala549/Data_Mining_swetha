```python

# 1)	Define the following metrics and perform the following operations
 i)	Write a Python program using Python Lists
 ii)	Write a Python program and NumPy

 Matrix A =  [[ 3.7827  3.3454  3.2341]  , [ 2.2122  3.5678  3.9087] , 
 [1.1234  2.8934,  5.9087]].

 Matrix B =  [[ 3.1234  3.0987  3.1234]  , [ 2.1111  3.2222  3.3333] , 
[1.0987  1.3456,  5.1234]].

 Matrix C =  [[ 3.1243  3.0989  3.1256 ]  , [ 2.6721  3.6785  3.9017] , 
 [1.1254  2.8956,  5.9187]].



import cProfile
import pstats

# Define matrices A, B, and C using Python lists
A = [
    [3.7827, 3.3454, 3.2341],
    [2.2122, 3.5678, 3.9087],
    [1.1234, 2.8934, 5.9087]
]

B = [
    [3.1234, 3.0987, 3.1234],
    [2.1111, 3.2222, 3.3333],
    [1.0987, 1.3456, 5.1234]
]

C = [
    [3.1243, 3.0989, 3.1256],
    [2.6721, 3.6785, 3.9017],
    [1.1254, 2.8956, 5.9187]
]

# Function to add two matrices element-wise
def elementwise_operation(matrix1, matrix2, operation):
    result_matrix = []
    for i in range(len(matrix1)):
        row = []
        for j in range(len(matrix1[0])):
            if operation == "add":
                row.append(matrix1[i][j] + matrix2[i][j])
            elif operation == "subtract":
                row.append(matrix1[i][j] - matrix2[i][j])
            elif operation == "multiply":
                row.append(matrix1[i][j] * matrix2[i][j])
        result_matrix.append(row)
    return result_matrix

def perform_operations():
    # Perform matrix operations
    sum_AB = elementwise_operation(A, B, "add")
    diff_AB = elementwise_operation(A, B, "subtract")
    prod_AB = elementwise_operation(A, B, "multiply")

    # Print results
    print("Matrix Addition A + B:", sum_AB)
    print("Matrix Subtraction A - B:", diff_AB)
    print("Matrix Element-wise Multiplication A * B:", prod_AB)

# Profile the perform_operations function
cProfile.run('perform_operations()', 'profile_list_stats')

# Print the profiling results
p = pstats.Stats('profile_list_stats')
p.strip_dirs().sort_stats('cumulative').print_stats(10)
```

```python
import numpy as np
import cProfile
import pstats

# Define matrices A, B, and C using NumPy arrays
A_np = np.array([
    [3.7827, 3.3454, 3.2341],
    [2.2122, 3.5678, 3.9087],
    [1.1234, 2.8934, 5.9087]
])

B_np = np.array([
    [3.1234, 3.0987, 3.1234],
    [2.1111, 3.2222, 3.3333],
    [1.0987, 1.3456, 5.1234]
])

C_np = np.array([
    [3.1243, 3.0989, 3.1256],
    [2.6721, 3.6785, 3.9017],
    [1.1254, 2.8956, 5.9187]
])

# Function to perform matrix operations using NumPy and operation type
def numpy_operation(matrix1, matrix2, operation):
    if operation == "add":
        return np.add(matrix1, matrix2)
    elif operation == "subtract":
        return np.subtract(matrix1, matrix2)
    elif operation == "multiply":
        return np.multiply(matrix1, matrix2)  # Element-wise multiplication

def perform_numpy_operations():
    # Perform matrix operations
    sum_AB_np = numpy_operation(A_np, B_np, "add")
    diff_AB_np = numpy_operation(A_np, B_np, "subtract")
    prod_AB_np = numpy_operation(A_np, B_np, "multiply")

    # Print results
    print("\nMatrix Addition A + B (NumPy):\n", sum_AB_np)
    print("\nMatrix Subtraction A - B (NumPy):\n", diff_AB_np)
    print("\nMatrix Element-wise Multiplication A * B (NumPy):\n", prod_AB_np)

# Profile the perform_numpy_operations function
cProfile.run('perform_numpy_operations()', 'profile_numpy_stats')

# Print the profiling results
p = pstats.Stats('profile_numpy_stats')
p.strip_dirs().sort_stats('cumulative').print_stats(10)
```

```python
# 2)	Write a Python Program and perform the following operations ?
I.	Create a List {1225, 4986, 6789, 7890, 2345, 6783, 0987, 1234, 8765, 3456}
II.	Iterate using a for loop
III.	Iterate using for loop and range
IV.	List Comprehension
V.	Enumerate
VI.	Iter function and next function
VII.	Map function
VIII.	Using zip
IX.	Using NumPy Module
```

```python
#normal 
import numpy as np
import cProfile
import pstats
from io import StringIO

# I. Create a List
swetha_list = [1225, 4986, 6789, 7890, 2345, 6783, 987, 1234, 8765, 3456]

# Function to iterate using a for loop
def iterate_for_loop():
    print("Iterating using a for loop:")
    for number in swetha_list:
        print(number)

# Function to iterate using a for loop and range
def iterate_for_loop_range():
    print("\nIterating using a for loop and range:")
    for i in range(len(swetha_list)):
        print(swetha_list[i])

# Function to perform list comprehension
def list_comprehension():
    print("\nList comprehension result (squares of the numbers):")
    squares = [x**2 for x in swetha_list]
    print(squares)

# Function to use enumerate
def use_enumerate():
    print("\nUsing enumerate:")
    for index, value in enumerate(swetha_list):
        print(f"Index {index}: Value {value}")

# Function to use iter() and next()
def use_iter_next():
    print("\nUsing iter() and next():")
    swetha_iter = iter(swetha_list)
    for _ in range(len(swetha_list)):
        print(next(swetha_iter))

# Function to use map()
def use_map():
    print("\nUsing map() function (multiply each element by 2):")
    doubled = list(map(lambda x: x * 2, swetha_list))
    print(doubled)

# Function to use zip
def use_zip():
    print("\nUsing zip (pairing each element with its index):")
    zipped = list(zip(range(len(swetha_list)), swetha_list))
    print(zipped)

# Function to use NumPy
def use_numpy():
    print("\nUsing NumPy Module (convert list to NumPy array and perform operations):")
    swetha_array = np.array(swetha_list)
    print("Original NumPy array:", swetha_array)
    print("Array + 10:", swetha_array + 10)
    print("Array squared:", swetha_array ** 2)

# Profiling the functions
def profile_code():
    # Profile the iterate_for_loop function
    print("Profile results for iterate_for_loop:")
    pr = cProfile.Profile()
    pr.enable()
    iterate_for_loop()
    pr.disable()
    s = StringIO()
    ps = pstats.Stats(pr, stream=s).sort_stats(pstats.SortKey.TIME)
    ps.print_stats()
    print(s.getvalue())

    # Profile the iterate_for_loop_range function
    print("Profile results for iterate_for_loop_range:")
    pr = cProfile.Profile()
    pr.enable()
    iterate_for_loop_range()
    pr.disable()
    s = StringIO()
    ps = pstats.Stats(pr, stream=s).sort_stats(pstats.SortKey.TIME)
    ps.print_stats()
    print(s.getvalue())

    # Profile the list_comprehension function
    print("Profile results for list_comprehension:")
    pr = cProfile.Profile()
    pr.enable()
    list_comprehension()
    pr.disable()
    s = StringIO()
    ps = pstats.Stats(pr, stream=s).sort_stats(pstats.SortKey.TIME)
    ps.print_stats()
    print(s.getvalue())

    # Profile the use_enumerate function
    print("Profile results for use_enumerate:")
    pr = cProfile.Profile()
    pr.enable()
    use_enumerate()
    pr.disable()
    s = StringIO()
    ps = pstats.Stats(pr, stream=s).sort_stats(pstats.SortKey.TIME)
    ps.print_stats()
    print(s.getvalue())

    # Profile the use_iter_next function
    print("Profile results for use_iter_next:")
    pr = cProfile.Profile()
    pr.enable()
    use_iter_next()
    pr.disable()
    s = StringIO()
    ps = pstats.Stats(pr, stream=s).sort_stats(pstats.SortKey.TIME)
    ps.print_stats()
    print(s.getvalue())

    # Profile the use_map function
    print("Profile results for use_map:")
    pr = cProfile.Profile()
    pr.enable()
    use_map()
    pr.disable()
    s = StringIO()
    ps = pstats.Stats(pr, stream=s).sort_stats(pstats.SortKey.TIME)
    ps.print_stats()
    print(s.getvalue())

    # Profile the use_zip function
    print("Profile results for use_zip:")
    pr = cProfile.Profile()
    pr.enable()
    use_zip()
    pr.disable()
    s = StringIO()
    ps = pstats.Stats(pr, stream=s).sort_stats(pstats.SortKey.TIME)
    ps.print_stats()
    print(s.getvalue())

    # Profile the use_numpy function
    print("Profile results for use_numpy:")
    pr = cProfile.Profile()
    pr.enable()
    use_numpy()
    pr.disable()
    s = StringIO()
    ps = pstats.Stats(pr, stream=s).sort_stats(pstats.SortKey.TIME)
    ps.print_stats()
    print(s.getvalue())

# Run the profiling
profile_code()

```

```python
# 3)	For a List A, B, C, D, E write a python program to compute all the combinations and permutations 
```

```python
import itertools
import cProfile

# Lists A, B, C, D, E
list_Alpha = 'A'
list_Beta = 'B'
list_Charlie = 'C'
list_Delta = 'D'
list_Echo = 'E'

# Combine all the lists (actually, the list names) into one list
combined_list = [list_Alpha, list_Beta, list_Charlie, list_Delta, list_Echo]

# Function to compute all combinations
def compute_combinations():
    print("\nAll combinations:")
    for combo_size in range(1, len(combined_list) + 1):
        combinations = list(itertools.combinations(combined_list, combo_size))
        print(f"Combinations of {combo_size} elements: {combinations}")

# Function to compute all permutations
def compute_permutations():
    print("\nAll permutations:")
    for perm_size in range(1, len(combined_list) + 1):
        permutations = list(itertools.permutations(combined_list, perm_size))
        print(f"Permutations of {perm_size} elements: {permutations}")

# Profiling the functions
print("Profile results for compute_combinations:")
cProfile.run('compute_combinations()')

print("\nProfile results for compute_permutations:")
cProfile.run('compute_permutations()')
```
